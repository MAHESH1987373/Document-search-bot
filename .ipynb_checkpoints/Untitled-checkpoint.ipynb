```python
# üì¶ Database setup
import sqlite3
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel

def get_db_connection():
    conn = sqlite3.connect('docbot_users.db')
    conn.row_factory = sqlite3.Row
    return conn

def initialize_db():
    conn = get_db_connection()
    cursor = conn.cursor()

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS admin_users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            email TEXT UNIQUE NOT NULL,
            password TEXT NOT NULL
        )
    ''')

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS regular_users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            email TEXT UNIQUE NOT NULL,
            password TEXT NOT NULL
        )
    ''')

    conn.commit()
    conn.close()

# Call this once to create tables
initialize_db()

```

```python

import fitz  # PyMuPDF for PDFs
import docx
import pandas as pd
import pytesseract
from pdf2image import convert_from_path
import pdb
import os

UPLOAD_DIR = "uploaded_files"
os.makedirs(UPLOAD_DIR, exist_ok=True)

documents = []


def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file using PyMuPDF (fitz). If no text is found, apply OCR."""
    text = ""

    try:
        doc = fitz.open(pdf_path)
        for page in doc:
            extracted = page.get_text("text")  # Extracts text properly
            if extracted:
                text += extracted + "\n"
        doc.close()

        if not text.strip():
            print("‚ö†Ô∏è No text extracted with PyMuPDF, using OCR...")
            return extract_text_with_ocr(pdf_path)  # Fallback to OCR

        return text.strip()
    
    except Exception as e:
        return f"‚ùå Error processing PDF: {str(e)}"


def extract_text_from_docx(docx_path):
    """Extract text from a DOCX file."""
    text = ""
    try:
        doc = docx.Document(docx_path)
        text = "\n".join([para.text for para in doc.paragraphs])
        
        if not text.strip():
            return "‚ö†Ô∏è No readable text extracted from the DOCX file."
        
        return text.strip()

    except Exception as e:
        return f"‚ùå Error processing DOCX: {str(e)}"


def extract_text_from_excel(excel_path):
    """Extract text from an Excel file."""
    try:
        df = pd.read_excel(excel_path, sheet_name=None)  # Read all sheets
        extracted_text = []

        for sheet, data in df.items():
            if not data.empty:
                extracted_text.append(f"üìÑ Sheet: {sheet}\n{data.to_string(index=False)}")
        
        return "\n\n".join(extracted_text) if extracted_text else "‚ö†Ô∏è No data found in the Excel file."

    except Exception as e:
        return f"‚ùå Error processing Excel file: {str(e)}"


        
def process_uploaded_file(file_path, file_type):
    """Extract text based on file type, executing only the required function."""
    try:
        if file_type == "pdf":
            return extract_text_from_pdf(file_path)  # ‚úÖ Only PDF function runs
        if file_type == "docx":
            return extract_text_from_docx(file_path)  # ‚úÖ Only DOCX function runs
        if file_type == "xlsx":
            return extract_text_from_excel(file_path)  # ‚úÖ Only Excel function runs
        
        return f"Unsupported file type: {file_type}"
    
    except Exception as e:
        return f"‚ùå Error processing file: {str(e)}"        
        

        


```

```python
from fastapi import FastAPI, UploadFile, File, Body  # ‚úÖ Import Body properly
import os
import shutil
import google.generativeai as genai
from fastapi import FastAPI, UploadFile, File
app = FastAPI()

documents = []  # Ensure this is defined globally

@app.post("/signup")
def signup(user: UserSignup):
    conn = get_db_connection()
    cursor = conn.cursor()
    
    try:
        if user.is_admin:
            cursor.execute("INSERT INTO admin_users (email, password) VALUES (?, ?)", (user.email, user.password))
        else:
            cursor.execute("INSERT INTO regular_users (email, password) VALUES (?, ?)", (user.email, user.password))
        conn.commit()
    except Exception as e:
        raise HTTPException(status_code=400, detail="Email already registered.")
    finally:
        conn.close()
    
    return {"message": "Signup successful."}

@app.post("/signin")
def signin(user: UserSignin):
    conn = get_db_connection()
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM admin_users WHERE email=? AND password=?", (user.email, user.password))
    admin = cursor.fetchone()

    cursor.execute("SELECT * FROM regular_users WHERE email=? AND password=?", (user.email, user.password))
    regular = cursor.fetchone()

    conn.close()

    if admin:
        return {"message": "Admin login successful.", "role": "admin"}
    elif regular:
        return {"message": "User login successful.", "role": "user"}
    else:
        raise HTTPException(status_code=401, detail="Invalid credentials.")


@app.post("/upload")
async def upload_file(file: UploadFile = File(...)):
    file_path = os.path.join(UPLOAD_DIR, file.filename)  # Save path
    
    # Save file
    with open(file_path, "wb") as f:
        f.write(await file.read())

    # Extract file extension
    file_extension = os.path.splitext(file.filename)[1].lower()
    file_type = file_extension.lstrip(".")  # Remove dot (".pdf" ‚Üí "pdf")

    try:
        # Read file content
        content = await file.read()
        text =process_uploaded_file(file_path,file_type)  # Ensure this function extracts actual text

        # Store actual extracted text
        documents.append({"filename": file.filename, "content": text})
     
        return {"message": "File uploaded successfully", "filename": file.filename}

    except Exception as e:
        return {"error": f"‚ùå File Upload Error: {str(e)}"}
        
      


```

```python

@app.post("/search")
async def search_query(query: str = Body(..., embed=True)):  
    """Search within uploaded documents using Google Gemini API."""
    try:
        if not documents:
            return {"error": "No documents uploaded. Please upload a document first."}

        # ‚úÖ Print to check document content before sending to AI
        # print("üîç Debug: Current Documents List:", documents)

        # ‚úÖ Prepare the context (limiting to 5000 characters)
        context = "\n".join([doc["content"] for doc in documents])[:5000]

        # ‚úÖ Print to verify the extracted context
        # print("extracted text goes like this", context)

        # ‚úÖ Use Google Gemini for AI-powered search
        model = genai.GenerativeModel("gemini-1.5-flash")  
        response = model.generate_content(f"Context: {context}\nUser Query: {query}")

        # ‚úÖ Extract the response content
        answer = response.text if response.text else "No response generated."

        # ‚úÖ Print to debug AI response
        # print("ü§ñ Debug: AI Response:", answer)

        # ‚úÖ Log query and response
        with open("query_logs.txt", "a") as log_file:
            log_file.write(f"Query: {query}\nResponse: {answer}\n\n")

        return {"query": query, "response": answer}

    except Exception as e:
        return {"error": f"‚ùå API Error: {str(e)}"}


```

```python
@app.get("/files")
async def get_uploaded_files():
    """Returns the list of uploaded files."""
    return {"files": [doc["filename"] for doc in documents]}
```

```python
@app.get("/logs")
async def get_logs():
    """Fetch logs of queries and responses."""
    if not os.path.exists("query_logs.txt"):
        return {"logs": []}

    with open("query_logs.txt", "r") as log_file:
        logs = log_file.readlines()
    
    return {"logs": logs}
```

```python
from fastapi.middleware.cors import CORSMiddleware

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # Allow all frontend requests
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)
```
